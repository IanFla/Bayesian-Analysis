```python
data1 = np.array([[16, 58], [9, 90], [10, 48], [13, 57], [19, 103],
                  [20, 57], [18, 86], [17, 112], [35, 273], [55, 64]])
data2 = np.array([[12, 113], [1, 18], [2, 14], [4, 44], [9, 208],
                  [7, 67], [9, 29], [8, 154]])

n1 = data1.sum(axis=1).mean()
n2 = data2.sum(axis=1).mean()
Y = data1[:, 0] / data1.sum(axis=1)
Z = data2[:, 0] / data2.sum(axis=1)

model1 = lambda theta, x: beta(a=theta * n1, b=(1-theta) * n1).pdf(x)
model2 = lambda theta, x: beta(a=theta * n2, b=(1-theta) * n2).pdf(x)

posterior1 = lambda theta: np.prod([model1(theta, y) for y in Y], axis=0)
posterior2 = lambda theta: np.prod([model2(theta, z) for z in Z], axis=0)

theta = np.linspace(1/100000, 1-1/100000, 100000)
p1 = posterior1(theta)
p2 = posterior2(theta)
thetay = np.random.choice(theta, size=1000, p=p1/p1.sum())
thetaz = np.random.choice(theta, size=1000, p=p2/p2.sum())

plt.hist(thetay, density=True)
plt.hist(thetaz, density=True)
plt.show()

plt.hist(thetay - thetaz, density=True)
plt.show()
```

```python
x = np.array([-0.86, -0.3, -0.05, 0.73])
y = np.array([0, 1, 3, 5])
model = lambda a, b, x, y: (expit(a + b * x) ** y) * ((1 - expit(a + b * x)) ** (5-y))
likelihood = lambda a, b: np.prod([model(a, b, x[i], y[i]) for i in range(4)])

Sigma = [[2**2, 10], [10, 10**2]]
prior = lambda a, b: mvnorm(mean=[0, 10], cov=Sigma).pdf([a, b])
posterior = lambda a, b: prior(a, b) * likelihood(a, b)

def grid_contour(fun, x, y):
    z = np.ones([y.size, x.size])
    for i, xx in enumerate(x):
        for j, yy in enumerate(y):
            z[j, i] = fun(xx, yy)
            
    plt.contourf(x, y, z)
    plt.show()
    return z
    
a = np.linspace(-2, 4, 200)
b = np.linspace(-3, 23, 300)
z_pri = grid_contour(prior, a, b)
z_lik = grid_contour(likelihood, a, b)
z_pos = grid_contour(posterior, a, b)
```

# Ex 12

```python
years = np.array([1976, 1977, 1978, 1979, 1980, 1981, 1982, 1983, 1984, 1985]) - 1980
accidents = np.array([24, 25, 31, 31, 22, 21, 26, 20, 16, 22])

X = np.array([np.ones_like(years), years]).T
coef = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(accidents)
sigma2 = np.sum((accidents - X.dot(coef))**2) / (years.size - 2)
Sigma = 9 * np.linalg.inv(X.T.dot(X)) * sigma2
print(coef[0] + (1986 - 1980) * coef[1])

prior = lambda a, b: mvnorm(mean=coef, cov=Sigma).pdf([a, b])
def likelihood(a, b):
    if any(a+b*years <= 0):
        return 0
    else:
        return np.prod(poisson.pmf(k=accidents, mu=a+b*years))
    
posterior = lambda a, b: prior(a, b) * likelihood(a, b)

rg = np.sqrt(np.diag(Sigma))
a = np.linspace(coef[0] - rg[0], coef[0] + rg[0], 400)
b = np.linspace(coef[1] - rg[1], coef[1] + rg[1], 300)
z_pri = grid_contour(prior, a, b)
z_lik = grid_contour(likelihood, a, b)
z_pos = grid_contour(posterior, a, b)
```

```python
aa, bb = np.meshgrid(a, b)
predicts = (aa + (1986 - 1980) * bb).flatten()
weights = (z_lik / z_lik.sum()).flatten()
plt.hist(predicts, weights=weights, bins=50)
plt.show()

m = np.sum(weights * predicts)
std = np.sqrt(np.sum(weights * (predicts - m) ** 2))
print(m, std)
```
